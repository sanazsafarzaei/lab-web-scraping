#### 2.1. List all language names and number of related articles in the order they appear in wikipedia.org.

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
````

````python
# your code here
import requests
from bs4 import BeautifulSoup

url = 'https://www.wikipedia.org/'
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')
languages = soup.find_all('a', class_='link-box')

language_names_and_counts = []
for language in languages:
    name = language.find('strong').text
    count = language.find('small').text
    language_names_and_counts.append((name, count))

print(language_names_and_counts)
````

#### 2.2. Display the top 10 languages by number of native speakers stored in a pandas dataframe.
Hint: After finding the correct table you want to analyse, you can use a nested **for** loop to find the elements row by row (check out the 'td' and 'tr' tags). <br>An easier way to do it is using pd.read_html(), check out documentation [here](https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.read_html.html).

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
# your code here
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'

tables = pd.read_html(url)

df = tables[0]

n_columns = len(df.columns)

new_column_names = ["Language", "Native Speakers (in millions)"]
new_column_names += [None] * (n_columns - len(new_column_names))

df.columns = new_column_names

df["Native Speakers (in millions)"] = df["Native Speakers (in millions)"].astype(str)

df["Native Speakers (in millions)"] = df["Native Speakers (in millions)"].str.replace(",", "")


df["Native Speakers (in millions)"] = df["Native Speakers (in millions)"].astype(float)


df.sort_values("Native Speakers (in millions)", ascending=False, inplace=True)

display(df.head(10))







````

````python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.imdb.com/chart/top'

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
rows = soup.find_all('tr')

data = []
for row in rows:
    try:
        movie_name = row.find('td', {'class': 'titleColumn'}).find('a').text
    except AttributeError:
        movie_name = None
    try:
        movie_release = row.find('td', {'class': 'titleColumn'}).find('span').text[1:-1]
    except AttributeError:
        movie_release = None
    try:
        movie_director = row.find('td', {'class': 'titleColumn'}).find('span', {'class': 'credit'}).text.strip().split('\n')[0].strip()[3:-1]
    except AttributeError:
        movie_director = None
    try:
        movie_star = row.find('td', {'class': 'titleColumn'}).find('span', {'class': 'credit'}).text.strip().split('\n')[2].strip()[3:]
    except AttributeError:
        movie_star = None
        
    data.append([movie_name, movie_release, movie_director, movie_star])

df = pd.DataFrame(data, columns=['Movie Name', 'Release', 'Director', 'Stars'])
display(df.head())

````

#### 3.1. Display the movie name, year and a brief summary of the top 10 random movies (IMDB) as a pandas dataframe.

````python
#This is the url you will scrape in this exercise
url = 'https://www.imdb.com/list/ls009796553/'
````

````python
# your code hereimport requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.imdb.com/list/ls009796553/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

data = []
for movie in soup.find_all('div', {'class': 'lister-item mode-detail'})[:10]:
    name = movie.find('h3', {'class': 'lister-item-header'}).find('a').text
    year = movie.find('span', {'class': 'lister-item-year text-muted unbold'}).text.strip('()')
    summary = movie.find('p', {'class': 'text-muted'}).text.strip()
    
    data.append([name, year, summary])
    
df = pd.DataFrame(data, columns=['Movie Name', 'Year', 'Summary'])
display(df)
````
